디자인 패턴은 갱 오브 포(Gang of Four - GoF: Design Patterns: Reusable Object-Oriented Software)로 유명한 책의 소개와 함께 소프트웨어 공학에 널리 보급되어 옴  
디자인 패턴은 개발 중에 자주 발생하는 일반적인 문제들을 어떻게 추상화하여 해결할 수 있는지 소개함  
이것들을 적절하게 사용하면 일반화된 디자인을 사용해 문제를 쉽게 해결할 수 있음  

이 장에서는 **가장 자주 사용되는 디자인 패턴을 살펴볼 것**  
- 디자인 패턴이 어떤 조건에서 적용할 수 있는 도구인가의 관점 X
- 어떻게 클린코드에 기여할 수 있는 지의 관점 O  

디자인 패턴을 구현한 코드를 제시한 다음, 패턴을 사용하지 않았을 경우와 비교해 어떻게 더 나은 코드가 되었는 지를 비교/분석 

파이썬의 동적인 특성으로 인해 많은 디자인 패턴이 정적인 언어들과는 구현산의 차이가 있음  
즉, 파이썬에서 디장니 패턴을 적용할 때는 몇 가지 특별한 주의사항이 있으며, 때문에 어떤 경우에는 디자인 패턴을 적용하는 것이 오히려 파이썬 스럽지 않은 코드를 만들게 될 수 있음을 주의  

### 이 장에서 다루는 내용  
- 일반적인 디자인 패턴  
- 파이썬에서는 유효하지 않은 디자인 패턴 그리고 반드시 따라야 하는 관용적인 대안
- 가장 일반적인 디자인 패턴에 대한 파이썬스러운 구현 방식  
- 훌륭한 추상화가 자연스럽게 패턴으로 이어지는 원리

# 파이썬에 디자인 패턴 적용 시 고려사항
객체 지향 디자인 패턴은 다양한 시나리오에서 해결책의 하나로 제시되는 모델 중 하나임  
이는 특정 프로그래밍 언어에 종속된 것이 아니라 객체가 애플리케이션에서 상호작용하는 방법에 관한 보다 일반적인 개념에 가까움  
언어마다 다양한 세부 구현을 가지고 있짐나 디자인 패턴의 본질은 변하지 않음  
디자인 패턴은 이론적으로 객체의 레이아웃을 표현하는 추상적인 개념임  

파이썬의 특성을 생각해보면 고전적인 디장니 패턴 중 일부는 실제로 필요하지 않음  
-> 파이썬은 이미 그러한 디자인 패턴을 내부적으로 구현하고 있기 때문  
파이썬 자체에서 내부적으로 해당 기능을 지원하기 때문에 그것들이 지원되는지 알아차리지 못함  

어떤 것들은 파이썬의 동적인 특성으로 인해 훨씬 간단하게 구현할 수 있으며 어떤 것들은 다른 플랫폼과 약간의 차이만 있을 분 실질적으로 거의 동일함  

파이썬에서 클린코드를 달성하기 위한 중요한 목표는 어떤 패턴들이 있고 어떻게 구현할 수 있는지를 파악  
-> 이미 파이썬에서 추상화 한 패턴 중 일부를 인지하고 이를 활용할 수 있는 방법 터득
e.g) 이터레이션 패턴을 직접 구현하려는 시도는 파이썬스럽지 X => 파이썬에 이미 내재되어있는 개념이기 때문

파이썬은 생성패턴(creational pattern)의 일부와 비슷한 것을 지원함  
파이썬에서 클래스는 일반적인 객체일 뿐이며 함수도 마찬가지임  
- 클래스는 파라미터로 전달하거나 데코레이팅하거나, 재할당 하는 것이 가능  
    - **즉, 객체에 어떤 종류의 커스터마이징을 하더라도 특별한 팩토리 클래스가 필요 없음**  
- 파이썬에는 객체를 생성하기 위한 특별한 구문이 없음 (new 키워드 사용 불필요)  
    - **대부분의 경우 간단한 함수 호출은 팩토리처럼 동작**

다른 패턴 역시 파이썬의 매직 메서드나 표준 라이브러리에서 지원하는 기능을 활용하면 보다 파이썬스럽게 구현 가능
